```python
from sklearn.feature_extraction.text import TfidfVectorizer
import _pickle as pickle
import pandas as pd
```

```python
goodreads_irony_f = open("goodreads_irony_edited.pickle", mode="rb")
twitter_irony_f = open("twitter_irony_all.pickle", mode="rb")
```

```python
goodreads_irony = pickle.load(goodreads_irony_f)
twitter_irony = pickle.load(twitter_irony_f)
goodreads_irony = [i.lower() for i in goodreads_irony]
twitter_irony = [j.lower() for j in twitter_irony]
```

```python
print("Goodreads dataset size: " + str(len(goodreads_irony)))
print("Twitter dataset size: " + str(len(twitter_irony)))
```

```python
combined_dataset = goodreads_irony + twitter_irony
len(combined_dataset)
```

```python
goodreads_tfidf = TfidfVectorizer(min_df=0.05, max_df=0.6, ngram_range=(1,4))
goodreads_features = goodreads_tfidf.fit_transform(goodreads_irony)
pd.DataFrame(goodreads_features.todense(), columns=goodreads_tfidf.get_feature_names())
```
